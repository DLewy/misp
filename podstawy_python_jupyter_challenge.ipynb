#Challenge 0.2 i 0.3

```python
zajecia = [["wf"],
           ["informatyka", "nauka czytania"],
           ["relax", "jedzenie"],
           ["mispy","mispy w domu"],
           ["piateczek", "praca domowa z mispow", "Ilovemispy", "mispy w czasie snu"]
          ]

dnityg= ["Pon","Wto","Sro","Czwar","Pio"]
dnitygnum = [0,1,2,3,4]

dni = []
dni.append(dnityg)
dni.append(dnitygnum)

#### challenge 0.2
# zdefiniuj funkcje ktora poda ile jest zajec danego dnia
# lista_zajec_z_dnia -> ile ich jest
def podaj_ilosc_zajec(lista_zajec,dni):
  out = list()
  for i in dni[1]:
    out.append([dni[0][i],len(lista_zajec[i])])
  return out

#### challenge 0.3
# napisz program ktory zdajdzie dzien z najwieksza iloscia mispow
def podaj_ilosc_mispow(lista_zajec,dni):
  out = list()
  for i in dni[1]:
    num = 0
    for j in lista_zajec[i]:
      if "misp" in j:
        num += 1
    out.append([dni[0][i],num])
  return out
```

###Output dla challenge 0.2:
  - ilość zajęć w każdym dniu
  - dni posortowane po najbardziej zajętych

```python
#podaj ilość zajęc w każdym dniu
print(podaj_ilosc_zajec(zajecia,dni))
# posortuj dni po najbardziej zajetych
print(sorted(podaj_ilosc_zajec(zajecia,dni),key=lambda dniilosczaj: dniilosczaj[1],reverse=True))
```

###Output dla challenge 0.3:
  - dzień z największą ilościa mispów

```python
#print(sorted(podaj_ilosc_mispow(zajecia,dni),key=lambda dniilosczaj: dniilosczaj[1],reverse=True)[0])
print(max(podaj_ilosc_mispow(zajecia,dni),key=lambda ilosc: ilosc[1]))
```
